```python
#Loading libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading csv file in a dataframe
```

```python
df= pd.read_csv('all_data.csv')
```

```python
#Checking dataframe head
```

```python
df.head()
```

The dataframe is composed of 4 variables :

- Country: Country of the person
- Year: The year of birth of the person
- Life expectancy at birth (years): the estimate life expectancy of the person
- GDP: Gross Domestic Product (the total value of goods and services produced by a country in a year)

```python
#Let's shorten "Life expectancy at birth (years)" variable name for something more suitable
```

```python
df.columns = ['Country','Year','Expectancy','GDP']
```

For this project we want to see if there is a relation between Life Expectancy and GDP.
We want also to check how it evolve during the time each years.

```python
sns.set_palette("colorblind")
sns.set_style("whitegrid")
plt.figure(figsize=(15,10))
sns.barplot(data=df, x='Year', y='GDP', hue='Country')
plt.title('GDP per Country and per Years')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df, x='Year', y='Expectancy', hue='Country')
plt.title('Expectancy per Country and per Years')
plt.show()
```

```python
#Let's have a look to the Life Expectancy distribution by countries
```

```python
plt.figure(figsize=(15,10))
sns.violinplot(data=df,x='Country',y='Expectancy')
plt.title('Life expectancy distribution by country')
plt.show()
```

At first look Chile, Germany and USA distribution seems pretty similar then China and Mexico and finally Zimbabwa

```python
#Let's separate data by countries
df_chile= df[df.Country == 'Chile']
df_china= df[df.Country == 'China']
df_germany= df[df.Country == 'Germany']
df_mexico= df[df.Country == 'Mexico']
df_usa= df[df.Country == 'United States of America']
df_zimbabwe= df[df.Country == 'Zimbabwe']
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(data=df_chile, x='GDP', y='Expectancy',size_norm=True)
sns.scatterplot(data=df_china, x='GDP', y='Expectancy',size_norm=True)
sns.scatterplot(data=df_germany, x='GDP', y='Expectancy',size_norm=True)
sns.scatterplot(data=df_mexico, x='GDP', y='Expectancy',size_norm=True)
sns.scatterplot(data=df_usa, x='GDP', y='Expectancy',size_norm=True)
sns.scatterplot(data=df_zimbabwe, x='GDP', y='Expectancy',size_norm=True)
plt.title('Expectancy distribution by GDP and by Country')
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe'])
plt.show()
```

Looking at this graph we can question ourself on the effect of GDP on Life Expectancy as despite having the highest GDP USA doesn't get better Life Expectancy than Chile nor Germany.
As well as lower GDP doesn't mean better Life Expectancy if we compare Chile to Zimbabwe.

To clarify this point let's do an hypothesis testing regarding GDP and Life Expectancy.

 - As null hypothesis H0 let's say that Life Expectancy is the name for each country despite their GDP
 
 - As testing hypothesis let's say that GDP as an effect on Life Expectancy
